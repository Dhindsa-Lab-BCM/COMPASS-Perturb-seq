```python
import scanpy as sc
import anndata as ad
import numpy as np
import pandas as pd
from cnmf import cNMF, Preprocess
from scipy.io import mmread
from scipy.sparse import csr_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading count matrix and metadata for iPSCs
ipsc_folder = "/mnt/mass_storage1/mass_storage_projects/compass/seurat_objects/iPSC_processed_sparse/"
counts = mmread(ipsc_folder + "2_integrated_Mixscale_filt_iPSC_counts.mtx").tocsr().T  #transpose to Cells x Genes
obs = pd.read_csv(ipsc_folder + "2_integrated_Mixscale_filt_iPSC_metadata.csv", index_col=0)
var = pd.read_csv(ipsc_folder + "2_integrated_Mixscale_filt_iPSC_genes.csv", index_col=1)

# Assemble and save
adata = sc.AnnData(X=counts, obs=obs, var=var)
#adata.write_h5ad(ipsc_folder + "cNMF_input_iPSC.h5ad")
adata
```

```python
meta = adata.obs
meta["sequencing_run"].unique()
```

```python
print(adata.obs['sequencing_run'].value_counts())
```

```python
k_range = [6, 8, 10, 12, 14, 16, 19, 23, 27, 30, 35, 40, 45, 50, 55, 60, 100]
print(len(k_range))
```

```python
cnmf_obj = cNMF(output_dir=ipsc_folder, name="iPSC_noBatchCorrection")
cnmf_obj.prepare(counts_fn=ipsc_folder + 'cNMF_input_iPSC.h5ad', 
                 components=k_range, #testing for optimal K
                 n_iter=20, 
                 seed=67)
cnmf_obj.factorize(worker_i=0, total_workers=1)
cnmf_obj.combine()
cnmf_obj.k_selection_plot()  #this generates a PNG in your output directory to help you pick K
```

```python
# Consensus clustering at optimal K
cnmf_obj.consensus(k=9, density_threshold=0.1, show_clustering=True, close_clustergram_fig=False)
```

```python
(usage, spectra_scores, spectra_tpm, top_genes) = cnmf_obj.load_results(K=9, density_threshold=0.1)
usage.head()
```

```python
usage_norm_forplot = usage.unstack().reset_index()
usage_norm_forplot
```

```python
usage_norm_forplot = usage.unstack().reset_index()
usage_norm_forplot = pd.merge(left=usage_norm_forplot, right=adata.obs, left_on='level_1', right_index=True)
usage_norm_forplot.columns = ['GEP', 'cell', 'Usage'] + list(adata.obs.columns)
usage_norm_forplot
```

```python
cluster_order = list(np.sort(usage_norm_forplot["target_gene"].unique()))

# Optional filter for COMPASS genes
compass_genes = ['KMT2A', 'KMT2B', 'KMT2C', 'KMT2E', 'SETD1A', 'SETD1B', 'NonTargeting']
cluster_order = [i for i in cluster_order if i in compass_genes]
cluster_order
```

```python
(fig,axes) = plt.subplots(len(cluster_order),1, figsize=(10,10), dpi=300, gridspec_kw={'hspace':.8})
for i,k in enumerate(cluster_order):
    g = usage_norm_forplot.loc[usage_norm_forplot['target_gene'] == k, :]
    sns.boxplot(x='GEP', y='Usage', hue='sequencing_run', ax=axes[i], data=g, fliersize=1)
    axes[i].set_title(k)
    axes[i].legend(bbox_to_anchor=(1,1))
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')
    if i != (len(cluster_order)-1):
        axes[i].set_xticklabels([])
        
    if i != 0:
        axes[i].legend().remove()
    else:
        axes[i].legend(bbox_to_anchor=(1,1))

#fig.savefig(ipsc_folder + "iPSC_cNMF_Usage_by_SequencingRun.png", bbox_inches='tight')
```
